```python

import pandas as pd
import numpy as np
```

```python
ss_kiosk_df = pd.read_excel("DATA SUPREME SURVEY FORM (Responses).xlsx")
ss_kiosk_rename = pd.read_excel("DATA_SUPREME_COLUMN_RENAME_MAPPING.xlsx")
ss_kiosk_drop = ss_kiosk_rename.loc[ss_kiosk_rename['column cleanse'] == 'Remove'].new.to_numpy()
```

```python
ss_kiosk01_df = ss_kiosk_df.copy()
df01_col_cleansed = ss_kiosk01_df.rename(columns=ss_kiosk_rename.set_index('old')['new'].to_dict()).drop(ss_kiosk_drop, axis=1)
df01_row_cleansed = df01_col_cleansed.loc[df01_col_cleansed['Kiosk Experience'] == 'Yes']
df01_row_cleansed
```
