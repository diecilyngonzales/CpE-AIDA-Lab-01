### Task 2: Getting Class Statistics
1. Create a function named `get_grades()` for computing the semestral grade of each student:

  `inputs`: `dictionary` of a class grade sheet.
  
  `outpus`: `dictionary` showing:
  * The prelim, midterms, and finals grades of each student
  * The semestral grade of each student

2. Create a function named `get_class_stats()`.
  
  `inputs`: `dictionary` from `get_grades()`.
  
  `outpus`: `dictionary` showing:
  * The lowest and highest prelim, midterm, finals, and semestral grades
  * the mean, median, mode, and standard deviation of the grades


```python
def get_grades(arr1):
  class_grades = {}
  sem_grades = []

  arr_keys = list(arr1.keys())

  for item in arr_keys:
    if item == 'name':
      continue
    class_grades[item] = arr1[item]
  
  for item in range(len(arr1[arr_keys[0]])):
    sem_grades.append(round(arr1[arr_keys[1]][item] * 0.3 + arr1[arr_keys[2]][item] * 0.3 + arr1[arr_keys[3]][item] * 0.4, 2))

  class_grades['sem_grades'] = sem_grades

  return class_grades
```

```python
class_grades = get_grades(class_dict)

pd.DataFrame.from_dict(class_grades)
```

```python
import numpy as np
from scipy import stats

def get_grade_stats(arr1):

  grade_stats = {}

  grade_stats['highest'] = max(arr1)
  grade_stats['lowest'] = min(arr1)
  grade_stats['mean'] = round(np.mean(arr1), 2)
  grade_stats['median'] = np.median(arr1)
  grade_stats['mode'] = stats.mode(arr1)

  return grade_stats
```

```python
def get_class_stats(arr1):
  class_stats = {}

  arr_keys = list(arr1.keys())

  for item in arr_keys:
    class_stats[item] = get_grade_stats(arr1[item])
  return class_stats
```

```python
class_stats = get_class_stats(class_grades)

pd.DataFrame.from_dict(class_stats)
```

  *Note*: The function should not contain iterative code blocks (i.e. `for` or `while` loops, or list comprehension)


```python
import numpy as np

def get_grades(arr1):
  class_grades = {}

  arr_keys = list(arr1.keys())

  class_grades = dict(arr1)
  del class_grades[arr_keys[0]]
  
  #sem_grade
  class_grades_mx = np.array(list(class_grades.values()))
  #multiplier
  sem_prct = np.array([0.3, 0.3, 0.4])

  class_grades['sem_grades'] = list(np.round_(sem_prct @ class_grades_mx, decimals = 4))

  return class_grades
```

```python
class Section:
  def __init__(self, class_sheet, class_lists, max_pop=30):
    self.max_pop = max_pop
    self.class_sheet = class_sheet
    self.class_lists = class_lists
    self.class_grades = self.__get_class_grades()

  def __get_class_grades(self):
    xl = pd.ExcelFile(self.class_sheet)

    df1 = xl.parse('class_grades')
    class_dict = df1.set_index('id').to_dict('list')
    
    return class_dict

  def __count_class(self):
    return len(self.class_grades)

  def __get_grade_stats(self, arr1):

    grade_stats = {}

    grade_stats['highest'] = max(arr1)
    grade_stats['lowest'] = min(arr1)
    grade_stats['mean'] = round(np.mean(arr1), 2)
    grade_stats['median'] = np.median(arr1)
    grade_stats['mode'] = stats.mode(arr1)

    return grade_stats

  def __get_sem_grades(self):
    arr_keys = list(self.get_grades().keys())
    sem_grades = np.array(self.get_grades()[arr_keys[3]])
    return sem_grades

  def get_similar_students(self):
    xl2 = pd.ExcelFile(self.class_lists)

    df2 = xl2.parse('class1')
    class1 = df2.to_dict('list')

    df3 = xl2.parse('class2', header=None)
    class2 = df3.to_dict('list')

    similar_students = list(class1.values())[0] + list(class2.values())[0]
    return sorted(list(set(similar_students)))
  
  def is_overloaded(self):
    is_overload = True if self.__count_class() > self.max_pop else False
    return is_overload
  
  #Added tasks
  def get_grades(self):
    dtclass_grades = {}

    arr_keys = list(self.class_grades.keys())

    dtclass_grades = dict(self.class_grades)
    del dtclass_grades[arr_keys[0]]
    
    #sem_grade
    class_grades_mx = np.array(list(dtclass_grades.values()))
    #multiplier
    sem_prct = np.array([0.3, 0.3, 0.4])

    dtclass_grades['sem_grades'] = list(np.round_(sem_prct @ class_grades_mx, decimals = 4))

    return dtclass_grades

  def get_class_stats(self):
    class_stats = {}

    arr_keys = list(self.get_grades().keys())

    for item in arr_keys:
      class_stats[item] = self.__get_grade_stats(self.get_grades()[item])
    return class_stats

  def get_failed(self):
    arr_keys = list(self.get_grades().keys())
    failures = np.array(self.__get_sem_grades())
    return list(failures[failures<=70])

  def fail_count(self):
    return len(self.get_failed())




#instantiation, Test Task 1
class_grades3 = Section("class_grades-1.xlsx", "classlists-1.xlsx")
```
