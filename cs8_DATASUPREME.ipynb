```python
import pandas as pd
import numpy as np
```

```python
ss_kiosk_df = pd.read_excel("DATA SUPREME SURVEY FORM (Responses).xlsx")
ss_kiosk_rename = pd.read_excel("DATA_SUPREME_COLUMN_RENAME_MAPPING.xlsx")
ss_kiosk_drop = ss_kiosk_rename.loc[ss_kiosk_rename['column cleanse'] == 'Remove'].new.to_numpy()
ss_kiosk_rename
```

```python
ss_kiosk01_df = ss_kiosk_df.copy()
df01_col_cleansed = ss_kiosk01_df.rename(columns=ss_kiosk_rename.set_index('old')['new'].to_dict()).drop(ss_kiosk_drop, axis=1)
df01_row_cleansed = df01_col_cleansed.loc[df01_col_cleansed['Kiosk Experience'] == 'Yes']
df01_row_cleansed
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="darkgrid")
```

```python
#Gets the old value of the column label in order for the title of the succeeding visualization make more sense
def get_old_name(new_name):
  return ss_kiosk_rename.loc[ss_kiosk_rename['new'] == new_name].old.to_numpy()[0]

def get_new_name(df_index):
  return ss_kiosk_rename.new.to_numpy()[df_index]

def sns_countplot(hue_value, data_value=df01_row_cleansed, x_value=get_new_name(43)):
  sns.countplot(x=x_value, hue=hue_value, data=data_value).set(title=get_old_name(hue_value))
  plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

def sns_catplot_count(hue_value, col_value, data_value=df01_row_cleansed, x_value=get_new_name(43)):
  ct_plot = sns.catplot(x=x_value, hue=hue_value, col=col_value, data=data_value, kind="count")
  ct_plot.set_titles("{col_name}")
```

```python
sns_countplot("Functional Completeness Binary")
```

```python
sns_catplot_count("Functional Completeness Binary", "Gender")
```
